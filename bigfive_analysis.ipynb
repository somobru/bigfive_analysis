```python
questoes = [
  [0, 1], [1, -1], [1, -1], [2, 1], [0, 1], [2, 1], [4, 1], [1, 1], [3, 1],
  [4, 1], [3, 1], [0, -1], [3, 1], [4, -1], [1, 1], [0, -1], [2, -1], [1, 1],
  [2, -1], [2, 1], [4, -1], [2, -1], [4, -1], [3, -1], [3, 1], [0, 1], [2, 1],
  [1, -1], [0, 1], [1, -1], [2, 1], [2, 1], [3, 1], [4, 1], [3, 1], [4, 1],
  [0, 1], [2, -1], [3, 1], [1, 1], [4, 1], [0, -1], [3, -1], [3, 1]
]
tracos = ['Extroversão', 'Afabilidade', 'Conscienciosidade', 'Abertura', 'Neuroticismo']

def calcular_escores(respostas):
    escores = [0] * 5
    contagem = [0] * 5
    i = 0
    for resposta in respostas:
        traco, polo = questoes[i]
        if polo == 1:
            pi = resposta - 1
        else:
            pi = 5 - resposta
        escores[traco] += pi
        contagem[traco] += 1
        i += 1
    media_escores = [escores[i] / contagem[i] if contagem[i] != 0 else 0 for i in range(5)]
    return media_escores

def classificar_tracos(media_escores):
    mais_acentuados = []
    polo_oposto = []
    medianos = []
    for i, escore in enumerate(media_escores):
        if escore > 2.5:
            mais_acentuados.append(tracos[i])
        elif escore < 1.5:
            polo_oposto.append(tracos[i])
        else:
            medianos.append(tracos[i])
    return mais_acentuados, polo_oposto, medianos

def calcula_tracos_faixa3():
    pontuacao_pessoas = []
    print("Insira as respostas do questionário.\nE insira '1' a qualquer momento para encerrar o programa.")
    print()
    contador=0
    while True:
        entrada_str = input("")
        contador = contador + 1
        contador_final = contador - 1
        if entrada_str == '1':
            break
        entrada = list(map(int, entrada_str.split(' ')))
        pontuacao_pessoas.append(entrada)
    print('Número de respostas:', contador_final)
    print()

    alto = []
    medio = []
    baixo = []

    for respostas in pontuacao_pessoas:
        media_escores = calcular_escores(respostas)
        mais_acentuados, polo_oposto, medianos = classificar_tracos(media_escores)
        alto.append(mais_acentuados)
        baixo.append(polo_oposto)
        medio.append(medianos)
    return alto, baixo, medio

def calcula_grupo():
    resultado_alto = []
    resultado_medio = []
    resultado_baixo = []
    alto, baixo, medio = calcula_tracos_faixa3()

    for sublista in alto:
        pessoa = []
        for t in sublista:
            if t == 'Extroversão':
                pessoa.append('ext_alto')
            elif t == 'Conscienciosidade':
                pessoa.append('csc_alto')
            elif t == 'Afabilidade':
                pessoa.append('afa_alto')
            elif t == 'Abertura':
                pessoa.append('abt_alto')
            elif t == 'Neuroticismo':
                pessoa.append('neuro_alto')
        resultado_alto.append(pessoa)

    for sublista in medio:
        pessoa = []
        for t in sublista:
            if t == 'Extroversão':
                pessoa.append('ext_medio')
            elif t == 'Conscienciosidade':
                pessoa.append('csc_medio')
            elif t == 'Afabilidade':
                pessoa.append('afa_medio')
            elif t == 'Abertura':
                pessoa.append('abt_medio')
            elif t == 'Neuroticismo':
                pessoa.append('neuro_medio')
        resultado_medio.append(pessoa)

    for sublista in baixo:
        pessoa = []
        for t in sublista:
            if t == 'Extroversão':
                pessoa.append('ext_baixo')
            elif t == 'Conscienciosidade':
                pessoa.append('csc_baixo')
            elif t == 'Afabilidade':
                pessoa.append('afa_baixo')
            elif t == 'Abertura':
                pessoa.append('abt_baixo')
            elif t == 'Neuroticismo':
                pessoa.append('neuro_baixo')
        resultado_baixo.append(pessoa)

    resultado = []
    for i in range(len(resultado_alto)):
        r_alto = resultado_alto[i]
        r_medio = resultado_medio[i]
        r_baixo = resultado_baixo[i]
        resultado.append(r_alto + r_medio + r_baixo)
    return resultado

def calcula_planejador(resultado):
    planejador_count = 0
    for pessoa in resultado:
      planejador = 'csc_alto' in pessoa and ('abt_medio' in pessoa or 'abt_baixo' in pessoa) and 'ext_baixo' in pessoa
      if planejador:
            planejador_count += 1
    return planejador_count

def calcula_analista(resultado):
    analista_count = 0
    for pessoa in resultado:
      analista = 'abt_alto' in pessoa and 'neuro_baixo' in pessoa and ('csc_alto' in pessoa or 'csc_baixo')
      if analista:
            analista_count += 1
    return analista_count

def calcula_mediador(resultado):
    mediador_count = 0
    for pessoa in resultado:
        mediador = 'afa_alto' in pessoa and 'ext_medio' in pessoa and 'neuro_baixo' in pessoa
        if mediador:
            mediador_count += 1
    return mediador_count

def calcula_lider(resultado):
    lider_count = 0
    for pessoa in resultado:
      lider = 'ext_alto' in pessoa  and 'csc_alto' in pessoa and 'neuro_baixo' in pessoa and ('afa_alto' in pessoa or 'afa_medio' in pessoa)
      if lider:
            lider_count += 1
    return lider_count

def calcular_numero_grupos_possiveis(resultado):
    lideres = calcula_lider(resultado)
    mediadores = calcula_mediador(resultado)
    analistas = calcula_analista(resultado)
    planejadores = calcula_planejador(resultado)


    if lideres > 0 and mediadores > 0 and analistas > 0 and planejadores > 0:
      if lideres <= mediadores and lideres <= analistas and lideres <= planejadores:
          return lideres
      elif mediadores <= lideres and mediadores <= analistas and mediadores <= planejadores:
            return mediadores
      elif analistas <= lideres and analistas <= mediadores and analistas <= planejadores:
        return analistas
      else:
        return planejadores
    else:
        return 0

def iniciar_menu():
    print('Seja bem-vindo ao organizador de grupos ideal!',
          '\n     - Selecione a opção 1 para iniciar o processamento',
          '\n     - Selecione a opção 0 para sair')
    print()

    entrada = int(input())
    if entrada == 1:
        resultado = calcula_grupo()
        numero_grupos = calcular_numero_grupos_possiveis(resultado)

        numero_lideres = calcula_lider(resultado)
        numero_mediadores = calcula_mediador(resultado)
        numero_analistas = calcula_analista(resultado)
        numero_planejadores = calcula_planejador(resultado)

        print('Número de líderes presentes:', numero_lideres)
        print('Número de mediadores presentes:', numero_mediadores)
        print('Número de analistas presentes:', numero_analistas)
        print('Número de planejadores presentes:', numero_planejadores)
        print()

        if numero_grupos > 0:
            print("Número máximo de grupos possíveis contendo pelo menos\numa pessoa de cada perfil:", numero_grupos)
            print()

        print('Obrigado por utilizar o organizador de grupos!')

    elif entrada == 0:
        print('Obrigado por utilizar o organizador de grupos!')

def main():
    iniciar_menu()


main()
```
